```python
# This notebook is for exploratory data analysis, initial setup, and experimenting with the modular code.
# It can be used to run the full pipeline interactively.

# Import necessary libraries
import os
import sys
sys.path.append('../src')  # Add src to path for imports
```

```python

# Install dependencies if needed (e.g., in Colab)
!pip install -r ../requirements.txt

# Clone and install nnU-Net with modifications
!git clone https://github.com/rixez/Brats21_KAIST_MRI_Lab.git
repo_path = "./Brats21_KAIST_MRI_Lab"

# Modify requirements_v2.txt to remove sklearn==0.0
req_file = os.path.join(repo_path, "requirements_v2.txt")
with open(req_file, "r") as f:
    lines = f.readlines()
lines = [line for line in lines if line.strip() != "sklearn==0.0"]
with open(req_file, "w") as f:
    f.writelines(lines)

# Modify setup.py to remove "sklearn"
setup_file = os.path.join(repo_path, "setup.py")
with open(setup_file, "r") as f:
    lines = f.readlines()
new_lines = []
for line in lines:
    if '"sklearn",' in line.strip():
        continue
    new_lines.append(line)
with open(setup_file, "w") as f:
    f.writelines(new_lines)

# Modify model_restore.py for torch.load
file_path_train = os.path.join(repo_path, "nnunet/training/model_restore.py")
with open(file_path_train, "r") as f:
    lines = f.readlines()
lines = [line.replace("all_params = [torch.load(i, map_location=torch.device('cpu')) for i in all_best_model_files]",
                     "all_params = [torch.load(i, map_location=torch.device('cpu'), weights_only=False) for i in all_best_model_files]")
         for line in lines]
with open(file_path_train, "w") as f:
    f.writelines(lines)

!pip uninstall nnunet -y
!pip install -r Brats21_KAIST_MRI_Lab/requirements_v2.txt
!cd Brats21_KAIST_MRI_Lab && pip install .

print("Setup complete. Now you can run training or inference from src/ files.")
```
